```python
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
df['month_year']
```

```python
df_trend = df.groupby('month_year').sum(numeric_only=True)['sales'].reset_index()
```

```python
#setting the figure size
plt.figure(figsize=(15,7))
plt.plot(df_trend['month_year'],df_trend['sales'], color = 'red')
plt.xticks(rotation = 'vertical', size = 8)
plt.show()
```

```python
# Grouping product name column
prod_sales = pd.DataFrame(df.groupby('product_name').sum(numeric_only = True)['sales'].reset_index().reset_index(drop=True))
prod_sales
```

```python
prod_sales
```

```python
prod_sales = prod_sales.sort_values('sales',ascending=False)
```

```python
prod_sales[:10]
```

```python
#which are the most selling products
```

```python
most_sell_prod = pd.DataFrame(df.groupby('product_name').sum(numeric_only = True)['quantity'].reset_index())
most_sell_prod
```

```python
most_sell_prod = most_sell_prod.sort_values('quantity', ascending = False)
```

```python
most_sell_prod[:10]
```
